```python
import pandas as pd
```

```python
proj_dir = "Z:/landscape_geoinfo/datacube"
source = "sentinel2"
fp = (
    f"{proj_dir}/working/nodata_stats/{source}/"
    f"nodata_calc_combined_{source}.csv"
)
df = pd.read_csv(fp)
```

```python
df
```

```python
# Describe index statistics across the full period
index_stats_full = (
    df.groupby("index")["percent_nodata"]
    .describe()
    .reset_index()
)
```

```python
index_stats_full
```

```python
index_stats_full.to_csv(
    (
        f"{proj_dir}/working/nodata_stats/"
        f"nodata_index_stats_full_{source}.csv"
    ),
    index=False
)
```

```python
# Describe index statistics across seasons
index_stats_seasons = (
    df.groupby(["index", "season"])["percent_nodata"]
    .describe()
    .reset_index()
)
```

```python
index_stats_seasons
```

```python
index_stats_seasons.to_csv(
    (
        f"{proj_dir}/working/nodata_stats/"
        f"nodata_index_stats_seasons_{source}.csv"
    ),
    index=False
)
```

```python
# Describe overall statistics across seasons
overall_stats_seasons = (
    df.groupby(["year", "season"])["percent_nodata"]
    .describe()
    .reset_index()
)
```

```python
overall_stats_seasons
```

```python
overall_stats_seasons.to_csv(
    (
        f"{proj_dir}/working/nodata_stats/"
        f"nodata_overall_stats_seasons_{source}.csv"
    ),
    index=False
)
```
